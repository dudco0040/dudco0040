```
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
            nums_map = {}  #숫자라는 키가 들어가면 인덱스가 나오도록 딕셔너리를 만든다.
            for i, nums in enumerate(nums):
                if target - nums in nums_map: #target에서 현재 num를 뺀 것이 nums map의 key에 있으면 
                    return[nums_map[target - nums], i] #현재의 num과 target에서 num을 뺀 배열을 return한다.
                nums_map[nums] = i 
```

```
def maxProfit(A):
	
    # A가 0또는 1의 값을 가지는 경우
    if len(A) < 2:
        return 0
	 
    min_price = A[0]   # 배열에 첫번 째 수 
    maxProfit = 0    # 최대 이익이 발생할 경우 저장 - 0 보다 작은 값일 경우 손해 -> 큰 값이 없으면 기본값이 0이 출력되도록
    no=0
    update_time_min_temp = 1  #업데이트가 일어나지 않은 경우, 첫번째 값이 최소인 경우
    update_time_min = 1
    for a in A: # 리스트를 하나씩 불러옴
        no+= 1
        print("날짜",no)
        c = a - min_price   #가장 첫번 째 수를 
        if maxProfit < c:
            maxProfit = c
            update_time_max_profit = no  #매도 시점 
            update_time_min = update_time_min_temp  #매수 시점
            #print("------------------------------------------------")
            #print("{}시점에 값이 업데이트 되었습니다.".format(no))  #최대 이익 prices[no], 매도 시점 no
            #print("------------------------------------------------")

            if maxProfit == 10000:  #최대 수익 발생으로 더 이상 값을 비교할 필요가 없음
                break

        if min_price > a:
            min_price = a
            update_time_min_temp = no
            print("min update:",update_time_min_temp,"  min_values:",min_price)

        #첫번째에 값이 최소값일 경우 예외처리
        #update_time_min_temp = no  

        #print(a,"maxProfit(output)=",maxProfit,"min_price=",min_price)

    #이익이 발생하지 않은 경우
    if maxProfit == 0:   
        print("최대 이익이 발생하지 않았습니다.")
    else:
        print(update_time_min,"에서 매수하여",update_time_max_profit,"에 매도하였을 때, ",maxProfit,"만큼 이익을 얻습니다. ")  # 판매 X 경우, 이 문장을 출력 불가능 변수가 정의가 안되버림 
    
    return maxProfit
```
